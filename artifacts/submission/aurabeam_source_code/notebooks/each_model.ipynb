Model 1: yolov5

```python
import yaml
from ultralytics import YOLO


yaml_path = f"{dataset.location}/data.yaml"

with open(yaml_path, 'r') as f:
    yaml_content = yaml.safe_load(f)


train_folder_path = f"{dataset.location}/train/images"
yaml_content['train'] = train_folder_path
yaml_content['val'] = train_folder_path
yaml_content['test'] = train_folder_path 

with open(yaml_path, 'w') as f:
    yaml.dump(yaml_content, f)

print("✅ Đã trỏ toàn bộ cấu hình về tập Train!")

# --- BƯỚC 2: LOAD MODEL VÀ CHẠY TEST ---
model = YOLO('YOUR_MODEL_PATH.pt')  # Thay YOUR_MODEL_PATH bằng đường dẫn thực tế đến model của bạn

print("🚀 Bắt đầu evaluate model trên chính tập Train...")
# Vẫn gọi split='test', nhưng thực chất YOLO sẽ đọc data trong thư mục train
metrics = model.val(
    data=yaml_path, 
    split='test', 
    imgsz=1536,       # ĐÂY LÀ CHÌA KHÓA: Phải khớp tuyệt đối với lúc train
    
    batch=16          # Có thể set batch để test nhanh hơn nếu RAM/VRAM cho phép
)

# --- BƯỚC 3: IN KẾT QUẢ ---
print("\n" + "="*30)
print("🎯 KẾT QUẢ EVALUATE (TRÊN TẬP TRAIN):")
print(f"mAP50-95 : {metrics.box.map:.4f}")    
print(f"mAP50    : {metrics.box.map50:.4f}")       
print(f"Precision: {metrics.box.mp:.4f}")      
print(f"Recall   : {metrics.box.mr:.4f}")         
print("="*30)
```
